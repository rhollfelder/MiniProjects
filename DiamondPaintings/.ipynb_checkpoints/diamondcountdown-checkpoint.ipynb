```python
def start():
    colors = int(input('Number of different colors: '))
    global diamonds
    diamonds = []
    for i in range(colors):
        id = i + 1
        diamonds.append(int(input("Diamonds for ID " + str(id) + ': ')))
    for j in range(len(diamonds)):
        print('\n' + 'ID' + str(j+1) + ": " + str((diamonds[j])))
    global initialnum
    initialnum = diamonds
```

```python
def done(id,num):
    global diamonds
    diamonds[id-1] -= int(num)
    print ('ID' + str(id) + ': '+ str(diamonds[id-1]))
```

Use start() to create list of initial # of diamonds

initialnum[] is saved for later to show how many have been done
